```python
for column in df.columns:
    print(f"\nAttribute: {column}")
    print("Unique Values:")
    unique_values = df[column].unique()
    nunique_values = df[column].nunique()
    print(unique_values)
    print(f"No. unique values: {nunique_values}")

    # For quantitative attributes, display the range
    if df[column].dtype in ['int64', 'float64']:
        print(f"Range: {df[column].min()} to {df[column].max()}")

    # Check for missing values
    print(f"Missing Values: {df[column].isnull().sum()}")
```
